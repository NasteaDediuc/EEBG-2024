#**Lab 6 Assignment: Producing Publication-Ready Figures**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

lab3_data = sns.load_dataset('titanic')
lab3_data.loc[(lab3_data['who'] == 'child') & (lab3_data['sex'] == 'male'), 'who'] = 'boy'
lab3_data.loc[(lab3_data['who'] == 'child') & (lab3_data['sex'] == 'female'), 'who'] = 'girl'
lab3_data.loc[lab3_data['survived'] == 0, 'survived_str'] = 'no'
lab3_data.loc[lab3_data['survived'] == 1,  'survived_str'] = 'yes'
```

```python
sns.set_style("white")
sns.set_context("paper")
```

```python
palette = {'man': 'blue', 'woman': 'tab:orange', 'boy': 'cornflowerblue', 'girl': 'sandybrown'}
```

```python
ax = sns.barplot(data=lab3_data, x='class', y='survived', hue='who', capsize=0.05, errorbar=('ci',90), palette=palette)
```

```python
lab3_data_filtered = lab3_data[lab3_data['class'] != 'Second']
lab3_data_filtered.loc[:, 'class'] = pd.Categorical(lab3_data_filtered['class'], categories=['First', 'Third'])
ax = sns.barplot(data=lab3_data_filtered, x='class', y='survived', hue='who', capsize=0.05, errorbar=('ci', 90), palette=palette)
ax.set(xlabel='Boarding Class', ylabel='Mean Survival Rate')
sns.despine()
```

```python
palette = {'white'}
sns.boxplot(data=lab3_data, x='survived_str', y='age', hue='who', palette=palette)
```

```python
palette = {'man': 'blue', 'woman': 'tab:orange', 'boy': 'cornflowerblue', 'girl': 'sandybrown'}
sns.stripplot(data=lab3_data, x='survived_str', y='age', hue='who', jitter=0.3, alpha=0.5, dodge='true', palette=palette)
```

```python
transparent_palette = {'white'}
axes2 = sns.boxplot(data=lab3_data, x='survived_str', y='age', hue='who', palette=transparent_palette)
sns.stripplot(data=lab3_data, x='survived_str', y='age', hue='who', jitter=0.3, alpha=0.5, dodge=True, palette=palette, ax=axes2)
handles, labels = axes2.get_legend_handles_labels()
n = len(palette)
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', label=label, markerfacecolor=palette[label], markersize=10) for label in palette]
axes2.legend(handles=legend_elements, title='who')
```

```python
sns.boxplot(data=lab3_data, x='survived_str', y='age', hue='who', palette=transparent_palette, zorder=0)

sns.stripplot(data=lab3_data, x='survived_str', y='age', hue='who', jitter=0.3, alpha=0.5, dodge=True, palette=palette, zorder=5, legend=False)

plt.xlabel('Survived')
plt.ylabel('Age')
from matplotlib.lines import Line2D
handles, labels = plt.gca().get_legend_handles_labels()
legend_elements = [Line2D([0], [0], marker='o', color='w', label=label, markerfacecolor=palette[label], markersize=10) for label in labels]
plt.legend(handles=legend_elements, labels=labels, title='who')
sns.despine()
```

```python
ax = sns.violinplot(data=lab3_data, x='survived_str', y='age', hue='who', palette=palette, inner='point')
plt.xlabel('Survived')
plt.ylabel('Age')
sns.despine()
```

```python
palette = {'man': 'blue', 'woman': 'tab:orange', 'boy': 'cornflowerblue', 'girl': 'sandybrown'}
sns.scatterplot(data=lab3_data, x='age', y='fare', hue='class')
plt.xlabel('Age')
plt.ylabel('Fare')
sns.despine()
```

```python
lab3_data_subset=lab3_data[['class','who','fare']]
lab3_data_subset=lab3_data_subset.dropna(how='any')
lab3_data_subsetAverage=lab3_data_subset.groupby(['class','who']).mean()
lab3_data_subsetAverage=lab3_data_subsetAverage.reset_index()
lab3_data_subsetAverage.head()
```
